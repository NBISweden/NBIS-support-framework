**Metadata**  
There are three levels of metadata:
- For notebook level: in the Jupyter Notebook Toolbar go to Edit -> Edit Notebook Metadata. **Edit this for every report**.
- For cell level: in the Jupyter Notebook Toolbar go to View -> Cell Toolbar -> Edit Metadata and a button will appear above each cell.
- For output level: using `IPython.display.display(obj,metadata={"ipub":{}})`, you can set metadata specific to a certain output. Options set at the output level will override options set at the cell level. For an example of this, see [Multiple Outputs from a Single Code Cell](https://ipypublish.readthedocs.io/en/latest/code_cells.html#multiple-outputs).
- See [the full metadata definitions](https://ipypublish.readthedocs.io/en/latest/metadata_tags.html) to learn how to define different metadata blocks for your cells.

**Outputting**  
To convert to pdf run 
- `nbpublish -pdf -lb Example.ipynb --pdf-debug --clear-files`
- `nbpublish -pdf -lb -f latex_ipypublish_nocode Example.ipynb --pdf-debug  --clear-files` #hides notebook code unless allowed (see below)  
This permits merging multiple `.ipynb` into a single `.pdf`, preserving the metadata from the first notebook. See [ipypublish](https://ipypublish.readthedocs.io/en/latest) for more info.

# Embed interactive HTML (like ipywidgets)

Interactive HTML was created using <a href="https://ipyvolume.readthedocs.io">ipyvolume</a>
and will render below in .html type outputs:

```python
import numpy as np
import ipyvolume.pylab as p3
p3.figure()
p3.figure()
u = np.linspace(-10, 10, 25)
x, y = np.meshgrid(u, u)
r = np.sqrt(x**2+y**2).flatten()
x = x.flatten()
y = y.flatten()
time = np.linspace(0, np.pi*2, 15)
z = np.array([(np.cos(r + t) * np.exp(-r/5)) for t in time])
color = np.array([[np.cos(r + t), 1-np.abs(z[i]), 0.1+z[i]*0] for i, t in enumerate(time)])
color = np.transpose(color, (0, 2, 1))
s = p3.scatter(x, z, y, color=color, size=(z+1), marker="sphere")
p3.ylim(-3,3)
p3.show()
p3.save('../html/embed.html')
```
